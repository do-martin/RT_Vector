```python
pdf_dir = '/content/drive/MyDrive/docling_pdfs'
inhalt = os.listdir(pdf_dir)
for datei in inhalt:
    print(datei)
```

```python
# Mit Docling verarbeiten

converter = DocumentConverter()
markdown_dir = Path(f"{pdf_dir}/markdown")
json_dir = Path(f"{pdf_dir}/json")

markdown_dir.mkdir(exist_ok=True)
json_dir.mkdir(exist_ok=True)

for i, pdf_path in enumerate(Path(pdf_dir).glob("*.pdf")):
    md_path = markdown_dir / f"{pdf_path.stem}.md"
    json_path = json_dir / f"{pdf_path.stem}.json"

    if md_path.exists() or json_path.exists():
      print(f"[{i+1}] ⏩ Übersprungen: {pdf_path.stem}") 
      continue
        
    try:
        result = converter.convert(str(pdf_path))

        if md_path.exists() or json_path.exists():
          print(f"[{i+1}] ⏩ Übersprungen: {pdf_path.stem}") 
          continue

        result.document.save_as_json(f"{json_dir}/{pdf_path.stem}.json")
        markdown = result.document.export_to_markdown()
        md_path.write_text(markdown)
        print(f"[{i+1}] ✅ {pdf_path.stem}")
    except Exception as e:
        print(f"[{i+1}] ❌ {pdf_path.stem}: {e}")
```
