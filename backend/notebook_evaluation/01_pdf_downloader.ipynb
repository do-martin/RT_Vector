```python
from datasets import load_dataset
import requests
from pathlib import Path
```

```python
ds = load_dataset("G4KMU/vectara_open_ragbench", "Open RAGBench", split="text_tables")
ds
```

```python
unique_doc_ids = set(ds["doc_id"])
print(f"Anzahl unique Dokumente: {len(unique_doc_ids)}")
```

```python
pdf_dir = Path("pdfs")
pdf_dir.mkdir(exist_ok=True)

for i, doc_id in enumerate(unique_doc_ids):
    pdf_path = pdf_dir / f"{doc_id}.pdf"
    if pdf_path.exists():
        continue
    
    url = f"https://arxiv.org/pdf/{doc_id}.pdf"
    r = requests.get(url)
    
    if r.status_code == 200:
        pdf_path.write_bytes(r.content)
        print(f"[{i+1}/{len(unique_doc_ids)}] ✅ {doc_id}")
    else:
        print(f"[{i+1}/{len(unique_doc_ids)}] ❌ {doc_id} - Status {r.status_code}")
    
    # ArXiv Rate-Limit
    import time
    time.sleep(3)
```
