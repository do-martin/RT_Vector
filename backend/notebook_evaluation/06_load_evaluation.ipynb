```python
import pandas as pd

loaded_df = pd.read_csv("chunking_detailed_results.csv")
print(loaded_df.head())

metrics_to_compare = [
    'answer_correctness', 
    'faithfulness', 
    'context_precision', 
    'context_recall'
]

comparison_df = loaded_df.pivot_table(
    index='user_input',     
    columns='method',     
    values=metrics_to_compare
)
print(comparison_df.round(3))

sorted_df = loaded_df.sort_values(by=['user_input', 'method'])
text_comparison = sorted_df[[
    'user_input', 
    'method', 
    'response', 
    'answer_correctness'
]]
display(text_comparison)
```
